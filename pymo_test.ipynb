```python
from pymo.parsers import BVHParser
from pymo.preprocessing import JointSelector, MocapParameterizer, ConstantsRemover, Numpyfier, RootTransformer

from sklearn.pipeline import Pipeline
```

```python
PATH_TO_BVH = './files/SL_tts_sentence_12_tr_10fps.bvh'
```

```python
p = BVHParser()
```

```python
data_all = [p.parse(PATH_TO_BVH)]
```

```python
data_all[0].values
```

```python
data_all[0].values['LeftArm_Xrotation']
```

```python
data_pipe = Pipeline([
    ('jtsel', JointSelector(['Spine','Spine1','Spine2','Spine3','Neck','Neck1','Head','RightShoulder', 'RightArm', 'RightForeArm', 'RightHand', 'LeftShoulder', 'LeftArm', 'LeftForeArm', 'LeftHand'], include_root=True)),
    ('param', MocapParameterizer(param_type='euler')), # other possibilities are: 'euler', 'quat', 'expmap', 'position', 'expmap2pos'
    ('cnst', ConstantsRemover()),
    #('np', Numpyfier())
    ])
```

```python
out_data = data_pipe.fit_transform(data_all)
```

```python
data_df = out_data[0].values
```

```python
data_df
```

```python
data_df.columns.tolist()
```

It seems that we have some surplus joints: Neck1, Spine1, Spine2, Spine3. Interesting observation is that the duplicate columns are joints with numbers in the name. 

For example: Neck1_Xrotation appears two times in the dataframe:

```python
data_df['Neck1_Xrotation']
```

```python
df = data_df.loc[:,~data_df.columns.duplicated()]
```
