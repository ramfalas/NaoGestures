```python
def csv_to_angleInterpolation_lists(path_to_csv, fps=1):
    frame_slice = 20/fps
    df = pd.read_csv(path_to_csv, sep=",", index_col=0)

    names = df.columns.tolist()
    angleList = [df[name].tolist()[::frame_slice] for name in names]
    timeList = [time_item.total_seconds()[::frame_slice] for time_item in df.index.tolist()]*len(names)

    return names, angleList, timeList
```
